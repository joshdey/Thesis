```python
from ESN import simple_ESN as sesn
import ESN
import ESNteachers as teach
import numpy as np
import aux
import matplotlib.pyplot as plt
```

```python
x, y, z = teach.Lorenz(400, 0.01)
xrun, yrun, zrun = teach.Lorenz(10000, 0.01)
yz = np.concatenate((y,z), axis=1)
xz = np.concatenate((x,z), axis=1)
xy = np.concatenate((x,y), axis=1)
linearcoef=[0.76159, 0.41997]

lor = sesn(N=400, K=1, L=2, binary_node=0,a=1.0)
W = lor.generate_W(rho=1.0, dens=0.25, cont=0)
Win=lor.generate_Win(dens=0.8)

trainz = lor.train_ESN(input_dat=z, teacher=xy, around=linearcoef, order='power', washout=100, mp=False, B=0.01)

testrunz = lor.run_ESN(input_dat=zrun,around=linearcoef, order='power',time=10000)

lorattractorz=lor.plot_attractor(name='testLorenzz')

zout = lor.outputs

zrun1, zrun2=np.hsplit(zout, 2)
zrms=[aux.RMS(zrun1, xrun), aux.RMS(zrun2, zrun)]
zrmse1=aux.RMS_over_t(zrun1, xrun)
zrmse2=aux.RMS_over_t(zrun2, yrun)
zrmse = [zrms, zrmse1, zrmse2]

zrmse
```
