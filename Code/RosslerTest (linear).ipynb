```python
from ESN import simple_ESN as sesn
import ESN
import ESNteachers as teach
import numpy as np
import aux
import matplotlib.pyplot as plt
```

```python
x, y, z = teach.Rossler(400, 0.01)
yz = np.concatenate((y,z), axis=1)
linearcoef=[0.76159, 0.41997]

ross = sesn(N=400, K=1, L=2, binary_node=0,a=1.0)
W = ross.generate_W(rho=1.0, dens=0.25, cont=0)
Win=ross.generate_Win(dens=0.8)

train = ross.train_ESN(input_dat=x, teacher=yz, around=linearcoef, order='power', washout=100, mp=False, B=0.01)

xrun, yrun, zrun = teach.Rossler(30000, 0.01)

testrun, yz = ross.run_ESN(input_dat=xrun,around=linearcoef, order='power',time=30000)

rossattractor=ross.plot_attractor(name='testRossler')

out = ross.outputs

run1, run2=np.hsplit(out, 2)
rms=[aux.RMS(run1, yrun), aux.RMS(run2, zrun)]
rmse1=aux.RMS_over_t(run1, yrun)
rmse2=aux.RMS_over_t(run2, zrun)
rmse = [rms, rmse1, rmse2]
```

```python
rmse
```

```python
plt.plot(rmse1, '.', markersize=1)
plt.show()
```
